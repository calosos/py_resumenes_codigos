La decisión de qué hacer con los NaNs depende mucho del contexto. Vamos a ver 3 cosas básicas que podemos realizar para limpiar estos datos indeseables:

Eliminar filas con NaNs

Eliminar columnas con NaNs

Llenar los NaNs con algún valor.

```python
#Eliminar filas con NAns 
df_tres_droes= df_tres.dropna() #Para guardar el ddf hay que asignalo a una variable
df_tres_droes
```

```python
#Eliminar finlas en los que todos losd dato  de la fila  sean NaN
#El valor default es how='any', que elimina las filas donde haya mínimo un NaN.
df_tres_droes = df_tres.dropna(how='all')
df_tres_droes
```

```python
df_tres['descuesnto']= np.nan
df_tres
```

```python
#ELiminar columnas con Nans
#ambién el valor default es how=any, pandas elimina todas las columnas donde haya mínimo un NaN,
df_tres.dropna(axis=1)
```

```python
df_tres.dropna(axis=1, how='all')
```

```python
#LLenar cvalores nulos con un vlaor
#PRimero eliminar las filas y columnas donde tododslos valores sean nan 
df_droped = df_tres.dropna(axis=0, how= 'all')
df_droped = df_droped.dropna(axis=1, how= 'all')
df_droped
```

```python
#Rellerna con 0 la columna 'productos_vendidos
df_droped['productos_vendidos'] = df_droped['productos_vendidos'].fillna(0)
df_droped
```

```python
#ELimar los Nans que quedasn
df_cleaned = df_droped.dropna()
df_cleaned 
```
